```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
rows = ['A','B','C','D']
columns = ['2014','2015','2016','2017']
datas = np.random.uniform(50,500,(4,4))

df = pd.DataFrame(datas,rows,columns)
series1 = df.mean(axis=1)
```

# 2.

```python
sns.barplot(x=series1.index,y=series1.values)
```

```python
sns.barplot(x=series1.values,y=series1.index)
```
