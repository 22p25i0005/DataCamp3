```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

# 1.

```python
x = np.arange(0,50)
print(x)
y = x**2
plt.figure(figsize=[5,5])
plt.title("Graph x**2")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
plt.show()
```

# 2.

```python
x = np.arange(0,50)
print(x)
y = np.sin(x)
plt.figure(figsize=[5,5])
plt.title("Graph sin")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,'m')
plt.show()
```

# 3.

```python
x = y**2
print(x)
y = np.arange(0,50)
plt.figure(figsize=[5,5])
plt.title("Graph x = y^2")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,'#009900')
plt.show()
```

# 4.

```python
x = np.arange(0,50)
print(x)
y = np.cos(x)
plt.figure(figsize=[5,5])
plt.title("Graph cos")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,'r',ls='--')
plt.show()
```

# 5.

```python
x = np.arange(0,10)
y1 = x**2
y2 = x**3
fig = plt.figure()
ax1 = fig.add_subplot(111,title='graph')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.plot(x,y1,'r',marker='o')
ax1.plot(x,y2,'b',marker='^')
```

# 6.

```python
x1 = np.arange(0,10)
y1 = x1**2
x2 = y2**2
y2 = np.arange(0,10)
fig = plt.figure()

ax1 = fig.add_subplot(111,title='graph')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax1.plot(x1,y1,'r',marker='o')
ax1.plot(x2,y2,'b',marker='^')

ax1.legend(('y1','y2'))
plt.show()
```

# 9.

```python
df = pd.read_csv('train.csv')
df
```

```python
pcount = df['Pclass'].value_counts()
display(pcount)
plist = pcount.tolist()
print(plist)
```

```python
plt.figure(figsize=[5,5])
labels = ['Pclass 3','Pclass 1','Pclass 2']
plt.pie(plist, labels=labels, autopct='%1.0f%%', textprops={'color':"w"})

plt.show()
```

# 10.

```python
ecount = df['Embarked'].value_counts()
display(ecount)
elist = ecount.tolist()
print(elist)
```

```python
plt.figure(figsize=[5,5])
elabels = ['Southampton','Cherbourg','Queenstown']
explode = [0,0,0.2]
plt.pie(elist, labels=elabels,explode=explode, autopct='%1.0f%%', textprops={'color':"w"})

plt.show()
```
