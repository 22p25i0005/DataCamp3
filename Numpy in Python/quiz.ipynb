```python
import time
import array as arr
import random


randomlist = []
for i in range(20):
    n = random.uniform(10,99)
    randomlist.append(n)
    
print('วิธี 1')
temp = []
time1 = time.time()
for x in randomlist:
    temp.append(x**2)

ar = arr.array('d',temp)
time2 = time.time()
print('{} มิลลิวินาที'.format((time2-time1)*1000))

print('วิธี 2')
time3 = time.time()
newlist = [x**2 for x in randomlist]
ar = arr.array('d',newlist)
time4 = time.time()
print('{} มิลลิวินาที'.format((time4-time3)*1000))
```

```python
import numpy as np

temp = []
def checkRepeat(number):
    if number>10 and number<100:
        
        if len(temp)==0:
            temp.append(number)
        else:
            index = 0
            for y in temp:
                
                if number != y:

                    if index == len(temp)-1:
                        temp.append(number)
                        break

                else:
                    print('เลขซ้ำ กรุณาลองใหม่')
                    break
                index = index+1
    else:
        print('เลขต้องอยู่ระหว่าง 10-100')

while len(temp)<9:
    x = int(input('ป้อนตัวเลข'))
    checkRepeat(x)
my_array = np.array(temp)
my_array = my_array.reshape((3,3))
print('matrix 3x3 \n',my_array)
my_array_transpose = my_array.T
print('transpose \n',my_array_transpose)
print('เลขตำแหน่ง 2,2 : ',my_array_transpose[2,2])
    
```
